```python
import numpy as np
```

```python
np.random.seed(42)  # pour reproductibilité
```

```python
# Exemple : vocabulaire simple
vocab = ["I", "love", "JY", "AI"]
vocab_size = len(vocab)
d_model = 8  # dimension des embeddings
```

```python
# Initialisation aléatoire des embeddings
embeddings = np.random.randn(vocab_size, d_model)
```

```python
# Mapping mot -> embedding
word_to_idx = {word: i for i, word in enumerate(vocab)}
```

```python
# Fonction utilitaire pour transformer une phrase en embeddings
def encode_sentence(sentence):
    """
    Converts a list of words into a matrix of embeddings
    sentence : list of str
    return : numpy array of shape (len(sentence), d_model)
    """
    idxs = [word_to_idx[word] for word in sentence]
    return embeddings[idxs]
```

```python
# Test
sentence = ["I", "love", "AI"]
encoded = encode_sentence(sentence)
print("Encoded sentence shape:", encoded.shape)
print(encoded)
```
