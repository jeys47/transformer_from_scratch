```python
import numpy as np
```

```python
class Embedding:
    def __init__(self, vocab_size, d_model):
        self.vocab_size = vocab_size
        self.d_model = d_model
        self.weights = np.random.randn(vocab_size, d_model) * .01

    def forward(self, x):
        # x : indices of shape (seq_len,)
        return self.weights[x]
```

```python
## TEST RAPIDE AVEC INDICES FICTIFS

vocab_size = 10
d_model = 16
x = np.array([1, 3, 2, 5])
embed = Embedding(vocab_size, d_model)
out = embed.forward(x)
print(out.shape)
```
